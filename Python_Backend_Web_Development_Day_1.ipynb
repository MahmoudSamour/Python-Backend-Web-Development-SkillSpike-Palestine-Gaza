### Variables, Data Types, and Operators
Variables are used to store information that can be referenced and manipulated in a program. Let's start by creating some variables and understanding their data types.

````python
# Creating variables
name = "John"
age = 25
height = 5.9
is_student = True

# Printing variables and their types
print(f"Name: {name}, Type: {type(name)}")
print(f"Age: {age}, Type: {type(age)}")
print(f"Height: {height}, Type: {type(height)}")
print(f"Is Student: {is_student}, Type: {type(is_student)}")
````

### Control Structures: Conditionals and Loops

Control structures allow us to execute certain blocks of code based on conditions or repeatedly execute a block of code. Let's explore some basic control structures in Python.

````python
# Conditional Statements
if age < 20:
    print("You are a teenager.")
elif age < 30:
    print("You are in your twenties.")
else:
    print("You are 30 or older.")

# Looping through a list
fruits = ['apple', 'banana', 'cherry']
for fruit in fruits:
    print(f"I like {fruit}.")
````

### Functions and Modules

Functions allow us to group a set of statements together to perform a specific task. Modules, on the other hand, are Python files that contain a set of functions and variables. Let's create a simple function and explore how to use modules.

````python
# Defining a function
def greet(name):
    return f"Hello, {name}!"

# Using the function
print(greet("John"))

# Using a module
import math
print(f"The square root of 16 is: {math.sqrt(16)}")
````

## Additional Material: PEP 8 and Code Style

### Quick Overview

PEP 8, sometimes spelled PEP8 or PEP-8, is the style guide for Python code, containing conventions for writing code in Python. It was written with the primary focus of improving the readability of code and making it consistent across the wide spectrum of Python code.

### Significance

1. **Readability**: Code is read more often than it is written. A style guide provides a consistent format for new code, ensuring that it is easy to read and understand.

2. **Consistency**: When code is consistent, it’s easier to modify, with fewer errors and reduced maintenance.

3. **Community**: Adopting a common style guide can make it easier for other Python developers to read and understand your code, and to understand what to expect in terms of code quality, layout, and structure.

### Key Points from PEP 8

- Use 4 spaces per indentation level.
- Limit all lines to a maximum of 79 characters for code, and 72 for docstrings.
- Use blank lines to separate functions, classes, and blocks of code inside functions.
- When possible, put comments on a line of their own.
- Use docstrings (string literals) for modules, functions, methods, and classes.
- Import standard libraries first, followed by third-party libraries, and then your own modules.
- Use lowercase with words separated by underscores for function and variable names (snake_case) and use CapWords for class names.

### Example

Here's a small example that adheres to PEP 8 guidelines:

```python
import math

def calculate_area(radius):
    """Calculate the area of a circle."""
    return math.pi * radius ** 2
```

### Note

While PEP 8 provides a set of guidelines, it's essential to remember that they are just that - guidelines. In some cases, it might make sense to deviate from them. The primary goal should always be readability and maintainability of the code.

## Docstrings in Python

In Python, a docstring is a string literal that occurs as the first statement in a module, function, class, or method definition. It is used for explaining the purpose of the function/method and how it should be used. Docstrings are associated with the object as their `__doc__` attribute.

### Why are docstrings important?

1. **Documentation**: They provide a convenient way of associating documentation with Python modules, functions, classes, and methods. An object's docstring is defined by including a string constant as the first statement in the object's definition.

2. **Readability**: They improve the readability of your code. Other developers can read the docstring to understand what your function does, even if they don't have access to the function's code.

3. **Tool Integration**: Various tools, like Sphinx, can automatically produce online or printed documentation from docstrings. Additionally, the built-in `help()` function can retrieve the docstrings and provide helpful documentation on functions.

### How to write a docstring?

A docstring is a multi-line string that is enclosed between triple quotes. The first line is a brief summary of the function's purpose. If you feel the need to elaborate on the function, you can add more descriptions after the first line.

### Example

Here's the `calculate_area` function with an added docstring:

````python
import math

def calculate_area(radius):
    """
    Calculate the area of a circle given its radius.

    Parameters:
    - radius (float): The radius of the circle.

    Returns:
    - float: The area of the circle.
    """
    return math.pi * radius ** 2

# Accessing the docstring
print(calculate_area.__doc__)
````
